```python
print("🤖: こんにちは！京都のおすすめスポットを紹介します。")
print("🤖: どこに興味がありますか？（金閣寺 / 清水寺 / 嵐山 / 伏見稲荷）")

while(1):
    try:
        # ユーザーに入力を求める
        place = input("あなた: ")

        # 入力内容によって返事を変える (if - elif - else)
        if place == "金閣寺":
            print("🤖: 金閣寺は晴れた日の午前中が、池に反射して綺麗ですよ！")

        elif place == "清水寺":
            print("🤖: 清水寺は坂道が多いので、歩きやすい靴で行きましょう。")

        # --- ここから下を改造してください ---
        # 「嵐山」や「伏見稲荷」など、別の場所が入力された時の反応を追加しよう
        # 例：「嵐山」 と入力された時に、 「竹林の小径が有名ですね。トロッコ列車もおすすめです。」と答える

        elif place == "______":
            print("🤖: __________________________")

        # ----------------------------------

        else:
            print(f"🤖: すみません、「{place}」の情報はまだ勉強中です...")

    # 案内終了
    except KeyboardInterrupt:
        break

print("🤖: 案内システム終了しました。")
```
